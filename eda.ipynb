```python
import pandas as pd
```

```python
df = pd.read_csv('dataaa.csv')
print(df.head())
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
df = df.dropna()  
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Rating'], kde=True)  
plt.show()

sns.pairplot(df)  
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Rating'], kde=True)  # Example distribution plot
plt.show()

sns.pairplot(df)  # Example of pairplot
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Genre', y='Rating', data=df)
plt.xticks(rotation=45)
plt.title('Rating by Genre')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Votes', y='Rating', data=df)
plt.title('Votes vs. Rating')
plt.show()
```

```python
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df.groupby('Year')['Rating'].mean().plot(kind='line', figsize=(10, 5))
plt.title('Average Rating Over Years')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df['Director'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Directors by Movie Count")
plt.xlabel("Director")
plt.ylabel("Number of Movies")
plt.xticks(rotation=45)
plt.show()
```
